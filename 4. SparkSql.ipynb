### FUNCTIONS

```python
%sql
CREATE OR REPLACE FUNCTION spark_sql.ecommerce.discount_price(p_price DECIMAL(10,2))
RETURNS DECIMAL(10,2)
LANGUAGE SQL
RETURN p_price * 0.90
```

```python
%sql
SELECT price_per_unit, spark_sql.ecommerce.discount_price(price_per_unit) as discounted_price FROM spark_sql.ecommerce.ecomm_orders
```

### TABLE FUNCTIONS - UDTF

```python
%sql
CREATE OR REPLACE FUNCTION spark_sql.ecommerce.catergories(p_category STRING)
RETURNS TABLE
LANGUAGE SQL
RETURN
(SELECT * FROM spark_sql.ecommerce.ecomm_orders WHERE product_category = p_category)
```

```python
%sql
select * from spark_sql.ecommerce.catergories('Home Decor')
```
