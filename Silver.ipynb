#### MERGE with PySpark

```python
if spark.catalog.tableExists('datamodelling_vasu.silver.silver_table'):
    pass
else:
    spark.sql('''
              CREATE TABLE IF NOT EXISTS datamodelling_vasu.Silver.silver_table 
              AS
              SELECT * FROM silver_SOURCE''')
```

#### Merge with SQL 

```python
%sql
MERGE INTO datamodelling_vasu.silver.silver_table as ST
USING silver_source AS SS
ON ST.order_id = SS.order_id
WHEN MATCHED THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *
```
