### ROW LEVEL SECURITY

###### 1. Creating Mapping Table

```python
%sql
create or replace table spark_sql.ecommerce.mapping_table
(
  payment_category string,
  email string
)
```

```python
%sql
SELECT * FROM spark_sql.ecommerce.ecomm_orders
```

```python
%sql
INSERT INTO spark_sql.ecommerce.mapping_table
VALUES
('Credit Card', 'alevasanth0@gmail.com'),
('Debit Card', 'alevasanth0@gmail.com'),
('PayPal', 'vasanth.ale0502@gmail.com'),
('UPI', 'vasanth.ale0502@gmail.com');
```

```python
%sql
select * from spark_sql.ecommerce.mapping_table
where email = current_user();
```

### converting mapping table into a Boolean

```python
%sql
SELECT EXISTS
(
  select * from spark_sql.ecommerce.mapping_table
  where email = current_user()
  and payment_category = 'Credit Card'
)
```

### Coverting into a boolean function

```python
%sql
Create or replace function spark_sql.ecommerce.rowlevel_security(p_payment string)
returns boolean
language sql
return
(
  exists
  (
    select * from spark_sql.ecommerce.mapping_table
    where email = current_user() and payment_category = p_payment
  )
)
```

```python
%sql
ALTER TABLE spark_sql.ecommerce.ecomm_orders
SET ROW FILTER spark_sql.ecommerce.rowlevel_security ON (payment_method)
```

```python
%sql
SELECT * FROM spark_sql.ecommerce.ecomm_orders;
```
