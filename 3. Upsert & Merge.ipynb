#### UPSERT - MERGE

```python
df = spark.read.table("spark_sql.ecommerce.ecomm_orders")
```

```python
df.display()
```

```python
df.createOrReplaceTempView("source_ecomm")
```

```python
spark.sql('''select * from {orders_temp}''', orders_temp = df).display()
```

```python
%sql
select * from source_ecomm;
```

```python
%sql
merge into spark_sql.ecommerce.ecomm_orders trg
using source_ecomm src
on src.order_id = trg.order_id
when matched then update set *
when not matched then insert *
```

```python
%sql
delete from spark_sql.ecommerce.ecomm_orders trg;
```

```python
%sql
select * from spark_sql.ecommerce.ecomm_orders;
```
