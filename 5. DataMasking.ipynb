### MASK function

```python
%sql
CREATE OR REPLACE FUNCTION spark_sql.ecommerce.dynamic_mask(p_user_id string)
RETURN
CASE WHEN is_account_group_member('admin_vasu') THEN p_user_id
ELSE '********'
END;
```

```python
%sql
ALTER TABLE spark_sql.ecommerce.ecomm_orders
ALTER COLUMN user_id SET MASK spark_sql.ecommerce.dynamic_mask;
```

```python
%sql
SELECT * FROM spark_sql.ecommerce.ecomm_orders
```

```python
%sql
ALTER TABLE spark_sql.ecommerce.ecomm_orders
ALTER COLUMN user_id SET MASK WITH FUNCTION = spark_sql.ecommerce.dynamic_mask(user_id);
```
